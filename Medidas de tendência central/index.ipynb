# **Medidas de tendência central**

Quando queremos fazer um resumo estatístico dos dados o uso das medidas de tendência central é uma optima idea para podermos explicar os nosso dados para nosso cliente porque nós representamos esse cojunto de dados num único valor

Importância na prática:

* ✅ **Tomada de decisão:** Empresas usam a média salarial para definir políticas de pagamento.
* ✅ **Ciência de dados e ML:** Ajuda na pré-análise e na modelagem dos dados.
* ✅ **Saúde e economia:** Médicos usam a mediana de idade para entender grupos de risco.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Média**
* **Média:** – A soma de todos os valores dividida pelo número total de observações.
    *  Útil para conjuntos de dados simétricos sem valores extremos.

```python
# Jogadores brasileiros da copa do mundo de 2018 com as sua idades
Jogadores = {
    "Nomes":["Alisson", "Ederson", "Thiago Silva", "Marquinhos", "Miranda", "Casemiro", "Paulinho", "Fernandinho", "Willian", "Fred", "Neymar", "Gabriel Jesus", "Douglas Costa", "Taison", "Philippe Coutinho", "Renato Augusto", "Marcelo", "Danilo", "Fagner", "Filipe Luís"],
    "Idades":[25, 24, 33, 24, 33, 26, 29, 33, 29, 25, 26, 21, 27, 30, 26, 30, 30, 26, 29, 32] 
}

df_Jogadores = pd.DataFrame(Jogadores)
df_Jogadores
```

```python
# Quero o nome do jogador mais novo e do mais velho
nome_mais_novo =df_Jogadores["Nomes"][df_Jogadores["Idades"].idxmin()]
nome_mais_velho = df_Jogadores["Nomes"][df_Jogadores["Idades"].idxmax()]

#Quero saber idade do jogador mais novo e do mais velho
idade_mais_novo = df_Jogadores["Idades"].min()
idade_mais_velho = df_Jogadores["Idades"].max()

print(f"O mais novo na saleção é {nome_mais_novo} com {idade_mais_novo} anos\nO mais velho na saleção é o {nome_mais_velho} com {idade_mais_velho} anos")
```

```python
# Media das idades
media = df_Jogadores["Idades"].mean()
print(f"A média das idades é {media}")
```

```python
# representando graficamente
sns.set_theme()
plt.hist(df_Jogadores.Idades, bins = 7, color='lightGreen', edgecolor='black', alpha=0.7)
plt.axvline(media, color = "Blue", label = f"A meida é {media}", linestyle = "--")
plt.xlabel('Idades')
plt.ylabel('Frequência')
plt.title('Distribuição das idades dos jogadores')
plt.legend()
plt.show()
```

# **Mediana**
A **mediana** consiste no valor central de um conjunto de dados. Para obter a mediana, devemos organizar os dados em ordem crescente e verificar qual é o valor que se encontra na posição central. Caso a quantidade de dados seja um número par, a mediana será a média dos dois valores centrais. Exemplos:

```python
# Primeira forma de achar a mediana
mediana = df_Jogadores["Idades"].median()
print(f"A mediana das idades é {mediana}")
```

```python
#Segunda forma de achar a mediana
import statistics as st
mediana = st.median(df_Jogadores.Idades)
print(f"A mediana das idades é {mediana} anos")
```

```python
#Terceira forma
#Organizar os dados
orden_vals = np.sort(df_Jogadores.Idades)
print(f"Dados organizados: {orden_vals}")
print(f"A mediana das idades é {np.median(orden_vals)} anos")
```

```python
plt.hist(df_Jogadores.Idades, bins = 7, color = "lightblue", alpha = 0.7, edgecolor = "black")
plt.axvline(mediana, color = "red", label = f"A mediana é {mediana} anos", linestyle = "--")
plt.legend()
plt.xlabel("Idades dos Jogadores")
plt.title('Distribuição das idades dos jogadores')
plt.ylabel("Frequência")
plt.show()
```

# **Moda**
A moda consiste no valor que mais aparece (o mais frequente) em um conjunto de dados. Sendo assim, devemos contar quantas vezes um mesmo elemento aparece no nosso conjunto de dados. O que mais aparece, portanto, é a moda. E se dois ou mais elementos aparecerem a mesma quantidade de vezes? Nesse caso, temos várias modas 🙂 . 

```python
# primeira forma
moda = st.mode(df_Jogadores.Idades)
print(f"A moda é {moda}")
```

```python
plt.hist(df_Jogadores.Idades, bins = 7, alpha = 0.5, color = "blue", edgecolor = "black", label = "Destribuição de frequência")
plt.axvline(moda, color = "cyan", linestyle = "--", label = f"moda = {moda}")
plt.legend()
plt.xlabel("Idades dos Jogadores")
plt.title('Distribuição das idades dos jogadores')
plt.ylabel("Frequência")
plt.show()
```
